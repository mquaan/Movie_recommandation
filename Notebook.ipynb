## **1. Data Collection** <a name="collect"></a>

Download movie dataset via API

```python
base_url = 'https://api.themoviedb.org/3/'
movie_endpoint = 'movie/top_rated'

movies_file_path= 'Data/movies.csv'
discover_url = f'{base_url}{movie_endpoint}?api_key={API_KEY}&page={1}'
print(discover_url)
page = 1
total_pages = 50
header = []

if not os.path.exists(movies_file_path):
    movies_data = []
    
    while page <= total_pages:
        discover_url = f'{base_url}{movie_endpoint}?api_key={API_KEY}&page={page}'
        print(discover_url)
        discover_response = requests.get(discover_url)
    
        if discover_response.status_code == 200:
            discover_data = discover_response.json()
    
            if total_pages == 1:
                total_pages = discover_data['total_pages']
    
            if page == 1:
                header = list(discover_data['results'][0].keys())
                movies_data.append(header)
    
            for movie in discover_data['results']:
                movies_data.append([movie[field] for field in header])
    
            page += 1
            
        else:
            print(f"Error: Failed to retrieve discovered movies. Status Code: {discover_response.status_code}")
            break
            
    with open(movies_file_path, 'w', newline='', encoding='utf-8') as movies_file:
        movies_writer = csv.writer(movies_file)
        movies_writer.writerows(movies_data)
else:
    print(f"The file {movies_file_path} already exists.")
```

### **Types of recommendation system**

To understand movie recommendation systems in a little depth we need to understand the two broad types of recommendation systems which are: `Content Based` and `Collaborative based`.

**1. Content Based**: Content based filtering actually find similary between two movies based on movie's actor, producer, genre, content and so on. Suppose User see GodFather-1 then content recommand sytem suggest user GodFather-2.

**2. Collaborative Based**: Collaborative filtering basically find the similarity between two user. Suppose you have two user, user-1 watch GodFather-1 and user-2 watch both (Godfather-1 & GodFather-2) then based on user 2 recomendation system suggest user-1 to Godfather-2.

>For this project we will focusing on `Content Based Recommendation System`, as we think using metadata like `genre_names`, `casts`, `title`, and `overview` can provide us a good deal of insight on understanding users preferences and help recommend movies or TV shows according to this.

### **Plot description based Recommender**

We'll calculate similarity scores between movies using their plot descriptions found in the overview feature of our dataset. Then, we'll provide movie recommendations based on these similarity scores. Now, let's examine the data.

```python
model_df['overview'].head(5)
```

Now we will create a TF-IDF (Term Frequency-Inverse Document Frequency) matrix for the `overview` column of the `model_df` with following steps: 

<center>
<img src="https://mungingdata.files.wordpress.com/2017/11/equation.png?w=430&h=336" />    
</center>

- First, we import the `TfidfVectorizer` class from the `sklearn.feature_extraction.text` module. This class is used to convert a collection of raw documents to a matrix of TF-IDF features.

- Then, it creates an instance of `TfidfVectorizer` named `tfidf`, with English stop words removed. Stop words are common words like `the`, `is`, `and`, etc. that do not carry much meaningful information.

- Finally, it uses the `fit_transform` method of `tfidf` to learn the vocabulary and idf, return term-document matrix. This matrix is stored in `tfidf_matrix`.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(model_df['overview'])

tfidf_matrix.shape
```

For Similarity Production, there are many more similarity matrix but we discussed two of them
- `Cosine Similarity`

<center>
<img src="https://sites.temple.edu/tudsc/files/2017/03/cosine-equation.png"/>
</center>

- `Euclidean Similarity`
<center>
<img src="https://chrisjmccormick.files.wordpress.com/2014/07/euclidean-distance.png"/>
</center>

Now we try to use `cosine similarity` to calculate a numeric quantity that denotes the similarity between two movies. We use the cosine similarity score since it is indepentdent of magnitude and is relatively easy and fast to calculate.

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

We are now in a good positon to define our recommendation function

```python
def get_overview_based_recommendations(movie):
    matching_movie = model_df[model_df['title'] == movie]
    if not matching_movie.empty:
        index = matching_movie.index[0]
        distances = sorted(list(enumerate(cosine_sim[index])), reverse=True, key=lambda x: x[1])
        for i in distances[1:6]:
            print(model_df.iloc[i[0]].title)
    else:
        print(f"No movie {movie} found in the dataset!")
```

Now the user can enter the movie title and get recommendations

### **Title, Genre and Cast Based Recommender**

Without a doubt, enhancing the metadata used by our recommender would lead to improved recommendations. In this section, we'll focus on building a recommender that takes into account the following factors: `actors`, associated `genres`, and movie `titles`.

Now let's prepare the `casts` and `genre_names` data for a cosine similarity calculation by converting the lists of cast members and genres into string. 

```python
model_df['casts'] = model_df['casts'] .apply(lambda x: ' '.join(x))
```

```python
model_df['genre_names'] = model_df['genre_names'] .apply(lambda x: ' '.join(x))
```

Now let's create our metadata `tags`, which is a string that contains all the metadata that we want to feed to our vectorizer (namely movie `title`, `actors`, and `genres`)

```python
model_df['tags'] = model_df['title'] + model_df['casts']+ model_df['genre_names']
model_df['tags']
```

The next steps are the same as what we did our plot description based recommender. One important diffent is that instead of `TF-IDF`, we opt for `CountVectorizer()`. This choice ensures we don't undervalue `actors/genres` who have been part of more movies, which makes more intuitive sense.

<center>
<img src="https://www.educative.io/api/edpresso/shot/5197621598617600/image/6596233398321152"/>
</center>

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

Next, we transforms the `tags` column into a matrix of token counts, and then converts that matrix into a NumPy array.

```python
vector = cv.fit_transform(model_df['tags']).toarray()
```

```python
vector.shape
```

Now we calculate the `cosine_similarity` matrix using count matrix `vector` above

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

Now, let use heatmap that represents the similarity between the top 20 movies. This visualization provide a quick and intuitive way to understand the the relationships between different movie. By looking at the heatmap, you can easily see which movies are similar to each other.

```python
N = 20 

subset_similarity = similarity[:N, :N]

subset_titles = model_df.iloc[:N]['title']

sns.heatmap(subset_similarity, cmap="YlGnBu", xticklabels=subset_titles, yticklabels=subset_titles)
plt.title(f"Movie Similarity Heatmap - Top {N} Movies")
plt.show()
```

For example, from this chart above you can draw some comments:
- The darker blue color between `The Godfather` and `The Godfather Part II` suggests a higher similarity score, indicating a strong thematic and narrative connection between these two films.
- In contrast, the if you see a lighter yellow color between two movies, such as `The Godfather` and `12 Angry Men`, it indicates a lower similarity score. This makes sense as these two movies are from completely different genres and have different themes and storylines.

Finally we define a function which will get recommendations based on the movie title entered by the user 

```python
def get_tags_based_recommendations_cosine(movie):
    matching_movie = model_df[model_df['title'] == movie]
    if not matching_movie.empty:
        index = matching_movie.index[0]
        distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])
        for i in distances[1:6]:
            print(model_df.iloc[i[0]].title)
    else:
        print(f"No movie {movie} found in the dataset!")
```

Now the user can enter the movie title and get recommendations

In the above function we used `Cosine similarity`. Let's try to calculate count matrix above with `Euclidean similarity`

```python
from sklearn.metrics.pairwise import euclidean_distances

similarity_euclidean = 1 / (1 + euclidean_distances(vector))

def get_tags_based_recommendations_euclidean(movie):
    matching_movie = model_df[model_df['title'] == movie]
    if not matching_movie.empty:
        index = matching_movie.index[0]
        distances = sorted(list(enumerate(similarity_euclidean[index])), reverse=True, key=lambda x: x[1])
        for i in distances[1:6]:
            print(model_df.iloc[i[0]].title)
    else:
        print(f"No movie {movie} found in the dataset!")
```
